```python
def get_freq_counts(df):
    df.sort_values(by='fn_calls', inplace=True)
    df_counts = df.groupby('fn_calls').count()
    
    df_counts['freq'] = df_counts.rel_err.cumsum() / len(df)
    N_evals = np.array(df_counts.index)
    Freq = np.array(df_counts['freq'])
    Freq[-1] = Freq[-2] # Because that's the solutions converged till last iteration
    
    return N_evals, Freq
```
